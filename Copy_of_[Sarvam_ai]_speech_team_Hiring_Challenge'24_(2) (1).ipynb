````python
'''
To run these much models in one go, you need a A100 GPU with 40GB RAM, I ran it local and saved files
Calling models in a dict to make it robust via for loop, saving gpu stats, time taken, probability and transcribed texts with output you needed
'''



from faster_whisper import WhisperModel
import time
import pickle

sample_output_list = []
combined_output_list=[]
gpu_stats=[]
transcribed_texts=[]
probabilities=[]

MODELS = {
    "tiny.en": "Systran/faster-whisper-tiny.en",
    "tiny": "Systran/faster-whisper-tiny",
    "base.en": "Systran/faster-whisper-base.en",
    "base": "Systran/faster-whisper-base",
    "small.en": "Systran/faster-whisper-small.en",
    "small": "Systran/faster-whisper-small",
    "medium.en": "Systran/faster-whisper-medium.en",
    "medium": "Systran/faster-whisper-medium",
    "large-v1": "Systran/faster-whisper-large-v1",
    "large-v2": "Systran/faster-whisper-large-v2",
    "large-v3": "Systran/faster-whisper-large-v3",
    "large": "Systran/faster-whisper-large-v3",
    "distil-large-v2": "Systran/faster-distil-whisper-large-v2",
    "distil-medium.en": "Systran/faster-distil-whisper-medium.en",
    "distil-small.en": "Systran/faster-distil-whisper-small.en",
    "distil-large-v3": "Systran/faster-distil-whisper-large-v3",
}



for model_size in MODELS.keys():
    model_size = model_size
    device = "cuda"
    model = WhisperModel(model_size, device=device, compute_type="float32")
    start_time = time.time()
    transcribed_text = ""
    segments, info = model.transcribe("/content/audio.mp3", beam_size=14)
    for segment in segments:
        transcribed_text += segment.text
        segment_dict = {
            "model": model_size,
            "chunk_id": len(sample_output_list) + 1,  # Assuming chunk_id starts from 1 and increments by 1
            "chunk_length": segment.end - segment.start,
            "text": segment.text,
            "start_time": segment.start,
            "end_time": segment.end
        }
        sample_output_list.append(segment_dict)

    combined_output_list.extend(sample_output_list)
    end_time = time.time()
    time_taken = end_time - start_time
    transcribed_texts = [[model_size, transcribed_text]]
    gpu_stats = [[model_size, time_taken]]
    probabilities = [[model_size,info.language_probability]]
    print(f"{model_size} inference done!")

    # Save lists to files for each iteration
    with open(f"transcribed_texts_{model_size}.pkl", "wb") as f:
        pickle.dump(transcribed_texts, f)

    with open(f"gpu_stats_{model_size}.pkl", "wb") as f:
        pickle.dump(gpu_stats, f)

    with open(f"probabilities_{model_size}.pkl", "wb") as f:
        pickle.dump(probabilities, f)

    del model  # Clear model variable to save memory

with open("combined_output_list.pkl", "wb") as f:
    pickle.dump(combined_output_list, f)

'''After this group pkl files of gpu stats in one folder, transcribed texts in another, probabilities in another
I am downloading these via my shared folder links via gdown, later will unzip them to show results'''
````

````python
import gradio as gr
from pytube import YouTube
from moviepy.editor import VideoFileClip
from faster_whisper import WhisperModel

def process_video(youtube_link, model_type, model_size):

    yt = YouTube(youtube_link)
    stream = yt.streams.get_highest_resolution()
    stream.download()

    video = VideoFileClip(stream.default_filename)
    audio = video.audio
    audio.write_audiofile('audio.mp3')

    model = WhisperModel(model_size, device="cuda", compute_type="float32")
    segments, _ = model.transcribe("audio.mp3", beam_size=5)

    for segment in segments:
        yield {
            "chunk_id": segment.id,
            "chunk_length": segment.end - segment.start,
            "text": segment.text,
            "start_time": segment.start,
            "end_time": segment.end
        }

youtube_link = gr.Textbox(label="YouTube Link")
model_type = gr.Dropdown(["faster-whisper"], label="Model Type")
faster_whisper_options = ["large", "medium", "small", "base", "tiny", "distil-large-v3"]
model_size = gr.Dropdown(faster_whisper_options, label="Model Size")
inputs = [youtube_link, model_type, model_size]

output = gr.Textbox(label="Transcribed Segments", type="text")

gr.Interface(process_video, inputs, output, title="YouTube Audio Transcription",
             description="Transcribe audio from YouTube videos using different models.").launch(debug = True)
````

````python
'''Importing necessary libraries for scrapping, making directories, regex, and soup to get content'''

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium import *
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
import requests
import re
from bs4 import BeautifulSoup
import os
import time
````

````python
'''
Initialising directories
'''

text_dir = "/content/text_files"
audio_dir = "/content/audio_files"
````

````python
def do_one_book(id):
  print('doing book ', id)
  os.makedirs(text_dir, exist_ok=True)
  os.makedirs(audio_dir, exist_ok=True)
  page_load_timeout = 50 #adding timeouts as when you scrape data its dynamically scraped, so we gotta wait for page to load fully
  implicit_wait_time = 40 #adding more wait time just in case
  chrome_options = Options() #have to enable plethora of options like that of to accept cookies, javascript, debugging pipeline
  chrome_options.add_argument("--accept-cookies")
  chrome_options.add_argument('--enable-javascript')
  chrome_options.add_argument('--remote-debugging-pipe')
  chrome_options.add_argument('--headless')
  chrome_options.add_argument('--no-sandbox')
  chrome_options.add_argument('--disable-dev-shm-usage')
  driver = webdriver.Chrome(options=chrome_options)
  driver.set_page_load_timeout(page_load_timeout)
  driver.implicitly_wait(implicit_wait_time)
  soup = BeautifulSoup(driver.page_source) #taking content in a good html from driver source
  num_chapters = info[id] #taking number of chapters from dictionary variable we created

  #Looping them one by one, to fetch texts,verses,audio files, and downloading via wget
  #one can do inspect element and soup find all to find elements to fetch data from
  for chapter in range(1, num_chapters + 1):
      chapter_url = f"https://live.bible.is/bible/HINWTC/{id}/{chapter}"
      try:
          driver.get(chapter_url)
          wait = WebDriverWait(driver, 150)
          wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "audio-player-background")))
          soup = BeautifulSoup(driver.page_source, "html.parser")

          text_content = ""
          verse_elements = soup.find_all("span", {"data-verseid": True})
          for verse in verse_elements:
              verse_text = verse.get_text(strip=True)
              text_content += verse_text.strip() + "\n"
          text_filename = os.path.join(text_dir, f"{id}_{chapter}.txt")
          with open(text_filename, "w", encoding="utf-8") as text_file:
              text_file.write(text_content)

          audio_player_backgrounds = soup.find_all("div", {"class": "audio-player-background"})
          for idx, audio_player_background in enumerate(audio_player_backgrounds, start=1):
              audio_players = audio_player_background.find_all("video", {"class": "audio-player"})
              for audio_player in audio_players:
                  if audio_player.has_attr("src"):
                      audio_url = audio_player["src"]
                      if audio_url:
                          audio_filename = os.path.join(audio_dir, f"{id}_{chapter}_{idx}.mp3")
                          download_file(audio_url,audio_filename)
      except Exception as e:
          print(f"Exception occurred while processing {chapter_url}: {e}")
          time.sleep(60)  # Pause for 60 seconds between chapters to handle timeout exceptions
          driver.close()
          driver = webdriver.Chrome(options=chrome_options)
          driver.set_page_load_timeout(page_load_timeout)
          driver.implicitly_wait(implicit_wait_time)

def download_file(url, filename):
    os.system(f"wget {url} -o {filename}")
````

# Exploratory Data Analysis

#### FIRST WE WILL START BY CALCULATING THE SNR(SOUND TO NOISE RATIO) AND DURATION OF THE AUDIO CLIPS

#### ----> SNR IS BEING CALCULATED TO UNDERSTAND THE RELATIONSHIP BETWEEN THE ACTUAL AUDIO AND THE NOISE BEHIND IT. THIS IS A VERY IMPORTANT FACTOR WHILE TRAINING STT OR TTS MODELS, AS MORE NOISE WOULD LEAD TO POOR TRAINING. WE NEED DATA THAT HAS REALLY LOW SNR SCORE AROUND 0.01 OR LESS

#### ----> DURATION OF THE AUDIO CLIPS IS BEING CALCULATED TO UNDERSTAND HOW MUCH ACTUAL AUDIO DATA DOES THE CLIP CONTAIN. LONGER AUDIO CLIPS LEAD TO MORE DATA FOR THE MODEL TO UNCOVER PATTERNS AS VERY SHORT AUDIO CLIPS LEAD TO MAKING THE MODEL HALLUCINATE

````python
import pandas as pd
import numpy as np
import ast
import librosa

#calculating sound to noise ratio , it helps to make STT robust

def calculate_duration_and_snr(row):
    try:
        audio_array = row['audio']
        sr = row['sampling_rate']

        # Calculate duration
        duration = librosa.get_duration(y=audio_array, sr=sr)

        # Calculate SNR
        mean_signal = np.mean(audio_array)
        std_signal = np.std(audio_array)
        snr = mean_signal / std_signal

        return duration, snr
    except Exception as e:
        print(f"Error processing row: {e}")
        return None, None

try:
    # Calculate duration and SNR for each row
    df['duration'], df['snr'] = zip(*df.apply(calculate_duration_and_snr, axis=1))
except:
    df['duration'] = ""
    df['snr'] = ""
````

#### FOR SOME BETTER VISUALIZATION, WE WILL ZOOM INTO THE AUDIO SEGMENT TO GET A BETTER VIEW OF WHAT IS GOING ON

````python
def envelop(y,rate,threshold):
    mask = []
    y = pd.Series(y).apply(np.abs)
    y_mean = y.rolling(window=int(rate/10),min_periods=1,center=True).mean()

    for mean in y_mean:
        if mean > threshold:
            mask.append(True)
        else:
            mask.append(False)

    return mask
````

#### HERE WE WILL CREATE AN ENVELOP MASK, WHICH IS A VERY IMPORTANT TECHNIQUE USED TO IDENTIFY AREAS OF LOW AMPLITUDE AND ACTIVITY. ENVLOPE MASKS HELP US IDENTIFY WHERE THE STOPS AND PAUSES IN AN AUDIO ARE. AND THESE STOPS AND PAUSES ARE VERY IMPORTANT WHILE TRAINING AN STT MODEL AS IT LETS THE MODEL UNDERSTAND DIFFERENT SEGMENTS

````python
mask = envelop(df['audio'][0][start_sample:end_sample], df['sampling_rate'][0], 0.02)

plt.figure(figsize=(10, 4))
plt.plot(df['audio'][0][start_sample:end_sample], color="red")
plt.plot(mask * df['audio'][0][start_sample:end_sample], color='lightgreen')
plt.title('Audio Signal with Envelope')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.show()
````

#### BY OVERLAPPING THE ENVELOPE MASK, WE CAN SEE THAT THERE ARE CERTAIN AREAS WHERE WE HAVE A PAUSE, OR SPECIFICALLY IN THIS AUDIO, ITS JUST AN INSTRUMENT PLAYING, WHICH IN OUR CASE WILL COUNT AS A STOP OR PAUSE AS THE SPEAKER'S AMPLITUTDE HAS DECREASED

````python
import librosa.display

D = librosa.amplitude_to_db(np.abs(librosa.stft(df['audio'][0][start_sample:end_sample])), ref=np.max)

plt.figure(figsize=(10, 4))
librosa.display.specshow(D, sr=df['sampling_rate'][0], x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.show()
````

#### HERE WE ARE USING MFCC FEATURES TO UNDERSTAND THE AUDIO BETTER. THIS HELPS US VISUALIZE VERY IMPORTANT FEATURES WHICH ARE LATER VERY USEFULL FOR TRAINING STT MODELS. IN THIS GRAPH EACH COLOUR REPRESENTS A CERTAIN ENTITIES. THIS HELPS IN SPEAKER RECOGNITION AND UNDERSTANDING THE CHANGE IN AUDIO. AS YOU CAN SEE THERE ARE MAJORLY JUST TWO COLOURS, BLUE REPRESENTING THE MUSICAL INTRSUMENTS (DARK RED BEING SOME OTHER INSTRUMENT IN THE BACKGROUND) AND THE PEACH REPRESENTING THE SPEAKER WHICH IS THE MAJOR PART OF THE AUDIO. THIS SHOWS THAT THERE ARE NOT MANY ENTITIES IN THE AUDIO AND NO SUDDEN CHANGES IN ENTITIES

````python
# FFT has two parts (magnitude and frequency)
def calc_fft(y, rate):
    n = len(y)
    freq = np.fft.rfftfreq(n, d=1/rate) # This goes to the x-axis with the spaces between the samples (d)
    Y = abs(np.fft.rfft(y)/n) # Generate the magnitude - This is the y-axis. Normalize the length of the signal by dividing by n
    return (Y, freq)
````

````python
def plot_signals(signals):
    fig, axes = plt.subplots(nrows=2, ncols=5, sharex=False,
                             sharey=True, figsize=(20,5))
    fig.suptitle('Time Series', size=16)
    i = 0
    for x in range(2):
        for y in range(5):
            axes[x,y].set_title(list(signals.keys())[i])
            axes[x,y].plot(list(signals.values())[i])
            axes[x,y].get_xaxis().set_visible(False)
            axes[x,y].get_yaxis().set_visible(False)
            i += 1

def plot_fft(fft):
    fig, axes = plt.subplots(nrows=2, ncols=5, sharex=False,
                             sharey=True, figsize=(20,5))
    fig.suptitle('Fourier Transforms', size=16)
    i = 0
    for x in range(2):
        for y in range(5):
            data = list(fft.values())[i]
            Y, freq = data[0], data[1]
            axes[x,y].set_title(list(fft.keys())[i])
            axes[x,y].plot(freq, Y)
            axes[x,y].get_xaxis().set_visible(False)
            axes[x,y].get_yaxis().set_visible(False)
            i += 1

def plot_fbank(fbank):
    fig, axes = plt.subplots(nrows=2, ncols=5, sharex=False,
                             sharey=True, figsize=(20,5))
    fig.suptitle('Filter Bank Coefficients', size=16)
    i = 0
    for x in range(2):
        for y in range(5):
            axes[x,y].set_title(list(fbank.keys())[i])
            axes[x,y].imshow(list(fbank.values())[i],
                    cmap='hot', interpolation='nearest')
            axes[x,y].get_xaxis().set_visible(False)
            axes[x,y].get_yaxis().set_visible(False)
            i += 1

def plot_mfccs(mfccs):
    fig, axes = plt.subplots(nrows=2, ncols=5, sharex=False,
                             sharey=True, figsize=(20,5))
    fig.suptitle('Mel Frequency Cepstrum Coefficients', size=16)
    i = 0
    for x in range(2):
        for y in range(5):
            axes[x,y].set_title(list(mfccs.keys())[i])
            axes[x,y].imshow(list(mfccs.values())[i],
                    cmap='hot', interpolation='nearest')
            axes[x,y].get_xaxis().set_visible(False)
            axes[x,y].get_yaxis().set_visible(False)
            i += 1

def calc_fft(y, rate):
    n = len(y)
    freq = np.fft.rfftfreq(n, d=1/rate) # This goes to the x-axis with the spaces between the samples (d)
    Y = abs(np.fft.rfft(y)/n) # Generate the magnitude - This is the y-axis. Normalize the length of the signal by dividing by n
    return (Y, freq)


def envelop(y,rate,threshold):
    mask = []
    y = pd.Series(y).apply(np.abs)
    y_mean = y.rolling(window=int(rate/10),min_periods=1,center=True).mean()

    for mean in y_mean:
        if mean > threshold:
            mask.append(True)
        else:
            mask.append(False)

    return mask
````
